# _Next Step: Prepping the Data_


```python
# group by full_nm column (to eliminate duplicate name entries)
pd.DataFrame(data.groupby('full_nm')[['cty', 'st', 'zip']].first()).reset_index().head()
```

```python
# create a dataframe using the method above
physicians = pd.DataFrame(data.groupby('full_nm')[['cty', 'st', 'zip']].first()).reset_index()
```

```python
# confirm the format
physicians.head()
```

Ok here is where we're going to diverge slightly from yesterday. Instead of moving right into the physician counts by state, let's explore `zip` and assess the best approach for our purposes. To keep things simple lets remove the +4 and if we want to later, we can revisit getting more nuanced later.

```python
# what is the data type in the zip column?
physicians['zip'].dtype
```

Type `O` means object, which is the technical term used for strings. Since we're dealing with strings, let's see how we can go about getting the correct output (i.e. the first 5 characters in each line).

```python
# show the format that we want zip to be in
for code in physicians['zip'].head():
    print(code[:5])
```

```python
# a slightly easier way to achieve the same result
physicians['zip'].str[:5].head(10)
```

The strategy we used directly above is the one we'll use to update the `zip` column in our physician data.

```python
# update zip column to first five characters
physicians['zip'] = physicians['zip'].str[:5]; physicians.head()
```

Now that we have `zip` in the correct 5-character format, let's go ahead and get the count by zip code.

```python
# groupby zip code and return the count for each
zip_codes = pd.DataFrame(physicians.groupby('zip')['full_nm'].count()).reset_index(); zip_codes.head()
```

```python
# load in zip-fip data
fips = pd.read_csv('ZIP-COUNTY-FIPS_2017-06.csv', dtype='str'); fips.head()
```

```python
# sort by ZIP and then reset the index
fips = fips.sort_values(by='ZIP').reset_index(drop=True)
```

```python
fips['ZIP'].unique()
```

```python
# create a dictionary assigning zip code to key and corresponding fip code 
for index, row in fips.head().iterrows():
    print(index, row['ZIP'], row['STCOUNTYFP'])
```

```python
zip_dictionary = {}

# create a dictionary assigning zip code to key and corresponding fip code 
for index, row in fips.iterrows():
    zip_dictionary[row['ZIP']] = row['STCOUNTYFP']
```

```python
# replace zip codes with fips
fip_codes = zip_codes.replace(zip_dictionary); fip_codes.head()
```

```python
# rename columns for clarity
fip_codes.columns = ['fip', 'count']; fip_codes.head()
```

```python
fips = fip_codes['fip'].tolist()
values = fip_codes['count'].tolist()
```

```python
import plotly.figure_factory as ff

import numpy as np
import pandas as pd

df_sample = pd.read_csv('https://raw.githubusercontent.com/plotly/datasets/master/minoritymajority.csv')
df_sample_r = df_sample[df_sample['STNAME'] == 'Florida']

values = df_sample_r['TOT_POP'].tolist()
fips = df_sample_r['FIPS'].tolist()

endpts = list(np.mgrid[min(values):max(values):4j])
colorscale = ["#030512","#1d1d3b","#323268","#3d4b94","#3e6ab0",
              "#4989bc","#60a7c7","#85c5d3","#b7e0e4","#eafcfd"]
fig = ff.create_choropleth(
    fips=fips, values=values, scope=['Florida'], show_state_data=True,
    colorscale=colorscale, binning_endpoints=endpts, round_legend_values=True,
    plot_bgcolor='rgb(229,229,229)',
    paper_bgcolor='rgb(229,229,229)',
    legend_title='Population by County',
    county_outline={'color': 'rgb(255,255,255)', 'width': 0.5},
    exponent_format=True,
)
fig.layout.template = None
fig.show()
```
