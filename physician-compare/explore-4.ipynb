## _Prepping the Data_

We can officially visualize geodata! Now we need to figure out how to visualize the physician count data from our original data set. We need to do some further exploration/wrangling/editing to get it into a visualizable format. 

```python
# group by full_nm column (to eliminate duplicate name entries)
pd.DataFrame(data.groupby('full_nm')[['cty', 'st', 'zip']].first()).reset_index().head()
```

```python
# create a dataframe using the method above
physicians = pd.DataFrame(data.groupby('full_nm')[['cty', 'st', 'zip']].first()).reset_index()
```

```python
# confirm the format
physicians.head()
```

Ok so I think the easiest route to go from here would be to first group by state (i.e. `st`) and return the counts for each state. Then we can try and plot the counts for each state.

```python
# group by state and return counts
physicians.groupby('st')['full_nm'].count()[:5]
```

Lets go ahead and assign this to a variable and then see how we can plot these numbers.

```python
# physician counts by state
state_physician_counts = physicians.groupby('st')['full_nm'].count()
```

```python
# format of the contiguous US we use for mapping
contiguous_usa.sort_values('state').reset_index(drop=True)[:5]
```

We have our counts, but first, we need to make a few small tweaks to it. Our `state_physician_counts` includes all the states; however, our data set for mapping the United States only includes the _contiguous_ United States, which excludes Alaska and Hawaii. We'll include these eventually, but I wanted to start as simple as possible, which is why I decided to exclude them from the mapping. 

Our next steps are going to make sure that:
1. We update `state_physician_counts` so that it has the same states as `contiguous_usa`
2. Sort the states in `contiguous_usa` so mapping the physician counts is _a lot_ easier

```python
# see if Alaska and Hawaii are in the counts
print('AK' in state_physician_counts)
print('HI' in state_physician_counts)
```

```python
# what is the shape of state_physician_counts?
state_physician_counts.shape
```

It looks like we have to drop a few more than expected. If you remember from one of our previous notebooks, US terroritories like Puerto Rico and the US Virgin Islands were included in this dataset. We are going to have to eliminate those as well.

```python
# loop through abbreviations for territories to see if they are included in counts
not_contiguous = ['DC', 'AS', 'GU', 'MP', 'PR', 'VI']

for code in not_contiguous:
    print(str(code) in state_physician_counts)
```

Here is a background of what each abbreviation represents:
- `AS`: American Samoa
- `GU`: Guam
- `MP`: Northern Mariana Islands
- `PR`: Puerto Rico
- `VI`: US Virgin Islands

As we can see in the `for` loop above, all but American Samoa was in `state_physician_counts`. Let's remove these observations, in addition to Alaska and Hawaii and then try to attach them onto `contiguous_usa`.

```python
# state abbreviations to drop
state_drop = ['AK', 'HI', 'GU', 'MP', 'PR', 'VI']

# drop above abbreviations
contiguous_physician_counts = state_physician_counts.drop(state_drop)

# what is the shape of the new counts?
contiguous_physician_counts.shape
```

```python
# print out first 10 observations of contiguous count
contiguous_physician_counts[:10]
```

Cool it looks like we're good to go here! Now let's sort by state for `contiguous_usa` and then we can attach the counts onto it.

```python
# sort by state
sorted_usa = contiguous_usa.sort_values(by='state').reset_index(drop=True); sorted_usa.head()
```

```python
# reset index to make multiple columns (and thus easier to work with)
contiguous_physician_counts = contiguous_physician_counts.reset_index()
```

```python
# rename count columns
contiguous_physician_counts.columns = ['state', 'count']
```

```python
# see what the first few rows look like
contiguous_physician_counts[:5]
```

Now the next step is going to be a little tedious: replacing the state abbreviations with the actual state name. If you notice states like Arizon and Arkansas are if different orders depending on which dataframe you're looking at. Because of this we'll replace them with the actual state name and then reorder it to ensure everything matches up.

```python
# create list of state names
state_names = list(sorted_usa['state'])
```

```python
# create list of state abbreviations
state_abbreviations = list(contiguous_physician_counts['state'])
```

```python
# loop through the two state lists
for name, abbreviation in zip(state_names, state_abbreviations):
    print(name, abbreviation)
```

As we can see there are more than a few mixups. I think the easiest approach (or at least the one that is coming to my mind first) is to create a dictionary, with the keys equal to the abbreviations and values equal to the state names. Luckily I found a CSV file that contained the both the names and corresponding abbreviations. Even better is that it's in alphabetical order according to state not abbreviation! 

```python
# load in states CSV
states = pd.read_csv('states.csv'); states.head()
```

```python
# gather index of Alaska and Hawaii
states[(states['State'] == 'Alaska') | (states['State'] == 'Hawaii')]
```

```python
# drop index 1, 11 from states dataframe
states = states.drop([1, 11])
```

```python
# create empty dictionary
state_dict = {}

# see what iterrows does
for index, row in states.head().iterrows():
    print(row['State'], row['Abbreviation'])
```

Awesome! What we'll do next is create a dictionary by iterating through each row and extracting the information we need from each one related to each state. 

```python
# assign all abbreviations as keys in state dictionary
for index, row in states.iterrows():
    # extract state and abbreviation
    state = row['State']
    abbreviation = row['Abbreviation']
    # create entry in our state dictionary
    state_dict[abbreviation] = state

# print out dictionary to make sure it worked correctly
state_dict
```

```python
# replace state abbreviations with state names in physician counts
contiguous_physician_counts.replace(state_dict)[:5]
```

It worked! Let's reassign this current format to the counts dataframe, sort it alphabetically, and then assign it to the dataframe for mapping. 

```python
# do the above and reassign
contiguous_physician_counts = contiguous_physician_counts.replace(state_dict)
```

```python
# sort counts alphabetically
contiguous_physician_counts = contiguous_physician_counts.sort_values('state').reset_index(drop=True); 
contiguous_physician_counts
```

```python
# attach counts as a new column to sorted_usa
sorted_usa['count'] = contiguous_physician_counts['count'].astype(int)
```

```python
sorted_usa
```
